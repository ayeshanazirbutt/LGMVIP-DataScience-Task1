**1) IMPORTING THE LIBRARIES**

```python
#Importing required libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

**2) LOADING THE DATASET:**

```python
columns = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width', 'Class_Labels']
data = pd.read_csv('iris.data', names = columns)
data.head()
```

**4) SEPARATING INPUT COLUMNS AND OUTPUT COLUMNS**

```python
df = data.values
X = df[:, 0:4]
Y = df[:, 4] 
print(X)
print(Y)
```

**5) SPLITTING THE DATA INTO TRAINING AND TESTING**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```

**MODEL 1 : SUPPORT VECTOR MACHINE ALGORITHM**

```python
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(X_train, y_train)
```

```python
prediction1 = model_svc.predict(X_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction1)*100)
```

**MODEL 2 : LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)
```

```python
prediction2 = model_lr.predict(X_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction2)*100)
```

**MODEL 3 : DECISION TREES CLASSIFIER**

```python
from sklearn.tree import DecisionTreeClassifier
model_dtc = DecisionTreeClassifier()
model_dtc.fit(X_train, y_train)
```

```python
prediction3 = model_dtc.predict(X_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction3)*100)
```

**MODEL 4 : K  Nearest Neighbour (KNN)**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

# Create the KNN classifier
model_knn = KNeighborsClassifier(n_neighbors=8)

# Train the model
model_knn.fit(X_train, y_train)

# Make predictions on the test set
prediction_knn = model_knn.predict(X_test)

# Generate the classification report for KNN
report_knn = classification_report(y_test, prediction_knn)
print("Classification Report for KNN:")
print(report_knn)

# Testing against manual data for KNN
X_new = np.array([[3, 2, 1, 0.2], [4.9, 2.2, 3.8, 1.1], [5.3, 2.5, 4.6, 1.9]])

# Make predictions on the manual data for KNN
prediction_manual_knn = model_knn.predict(X_new)
print("Prediction of Species for KNN:", prediction_manual_knn)
```
